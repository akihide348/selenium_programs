```python
import urllib.request
from bs4 import BeautifulSoup


class Scraper:
    def __init__(self, site):
        self.site = site
        
    def scrape(self):
        r = urllib.request.urlopen(self.site)
        html = r.read()
        parser = "html.parser"
        sp = BeautifulSoup(html, parser)
        
        for tag in sp.select("a"):
            url = tag.get("href")
            
            
            title = tag.text
            int_title = len(tag.text)

            if int_title < 10:
                continue
            else:
                print(title)
        

        
            if url is None:
                continue
            
            if 'http' in url:
                print("\n" + url)
    
           
```

```python
news = 'https://b.hatena.ne.jp/hotentry/it'
Scraper(news).scrape()
```
